### Data Visualisation and Analysis

In this section, we will analyze the OECD DAC statistics of the G7 countries, which we imported and cleaned above, within the framework of financing flows in the fields of climate change mitigation and adaptation. Based on principal and significant flows in our analysis, we will try to make inferences about the willingness levels of countries to commit to provide $100 billion annually in funding. All figures subject to analysis are 2019 US Dollar constant prices and are expressed in million USD.

```python
#Enabling style selection for visualisation
import matplotlib.style as style
pd.options.display.float_format = "{:,.2f}".format
```

```python
#Looking G7 Principal Mitigation Finance Flows by Years
G7byyearsmitpr = resultfinal.groupby('Year').Principal_Mitigation.sum()
G7byyearsmitpr.head(10)
```

```python
#Looking G7 Principal Adaptation Finance Flows by Years
G7byyearsadapr = resultfinal.groupby('Year').Principal_Adaptation.sum()
G7byyearsadapr.head(10)
```

```python
#Looking G7 Principal Mitigation Finance Flows by Years
G7byyearsmitsig = resultfinal.groupby('Year').Significant_Mitigation.sum()
G7byyearsmitsig.head(10)
```

```python
#Looking G7 Principal Adaptation Finance Flows by Years
G7byyearsadasig = resultfinal.groupby('Year').Significant_Adaptation.sum()
G7byyearsadasig.head(10)
```

```python
#Selecting Style
style.use('fivethirtyeight')

#Creating a bar chart for comparing Principal Mitigation and Adaptation Finance Flows
adamittogetherpr = pd.DataFrame({"Pr. Mitigation" : G7byyearsmitpr, "Pr. Adaptation" : G7byyearsadapr})
adamittogetherprgraph = adamittogetherpr.plot.bar(rot=0, figsize=(12,8)).set(title='Comparaison between "Principal" Mitigation \n and \n Adaptation Flows (million USD)')

#Showing the graph and Data Features
plt.show()
adamittogetherpr.describe()
```

We couldn't observe a clear  trend for principal mitigation flows. On the other hand for principal adaptation flows, we see an upward trend compared to 2010 level but it is not a relentless increase due to the recess in 2018.

Between 2010 and 2019, while the maximum level of principal mitigation flow is 10.7 billion USD and the minimum level is 5.6 billion USD, even the max level in principal adaptation which is 3.6 billion USD is below the lowest mitigation flow. 

```python
#Selecting Style
style.use('fivethirtyeight')

#Creating a bar chart for showing the ration of 'Principal' Adaptation Finance Flows to total 'Principal' Finance Flows
adaprtototal = G7byyearsadapr / (G7byyearsadapr + G7byyearsmitpr)
adaprtotatalg7 = adaprtototal.plot.bar(x='Year', y='Adaptation to Total', figsize= (9, 3)).set(title='"Principal" Adaptation to Total "Principal" Flows')
```

With another saying,  principal adaptation finance flows to total principal flows ratio is well below %50 for the all period. Since 2015, we are seeing an upward pattern but it is still missing the commitments of 50-50 balanced distribution of funds among adaptation and mitigation. 

##### Same Analysis for "Significant" Finance Flows

```python
#Looking G7 Significant Mitigation and Adaptation Finance Flows by Years
G7byyearsmitsig = resultfinal.groupby('Year').Significant_Mitigation.sum()
G7byyearsadasig = resultfinal.groupby('Year').Significant_Adaptation.sum()

print(G7byyearsmitsig.head(10))
print(G7byyearsadasig.head(10))
```

```python
#Selecting Style
style.use('fivethirtyeight')

#Creating a bar chart for comparing "Significant" Mitigation and Adaptation Finance Flows
adamittogethersig = pd.DataFrame({"Sg. Mitigation" : G7byyearsmitsig, "Sg. Adaptation" : G7byyearsadasig})
adamittogethersiggraph = adamittogethersig.plot.bar(rot=0, figsize=(12,8)).set(title='Comparaison between "Significant" Mitigation \n and \n Adaptation Flows (million USD)')

#Showing the graph and Data Features
plt.show()
adamittogethersig.describe()
```

Despite the downbeat outlook of principal  flows, this is not the case for finance flows categorized as significant. We  are able to mention about a positive tend both adaptation and mitigation flows by and largely especially after 2015, we are se ing significant increase in both flows.  another interesting statistic in this period is, the mean values of both flows are very close to each other. This might be interpreted as G7 countries are successful to achieve balanced (50-50) finance flows in terms of significant adaptation and mitigation flows.

```python
#Selecting Style
style.use('fivethirtyeight')

#Creating a bar chart for showing the ration of 'Significant' Adaptation Finance Flows to total 'Significant' Finance Flows
adasigtototal = G7byyearsadasig / (G7byyearsadasig + G7byyearsmitsig)
adasigtotatalg7 = adasigtototal.plot.bar(x='Year', y='Adaptation to Total', figsize= (9, 3)).set(title='"Significant" Adaptation to Total "Significant" Flows')
```

As it can be seen in the graph above the ratio of significant adaptation flows to total significant flows was exceeding 50% levels before 2015, this ratio decrease in the Paris Agreement period to 50% and even lower levels.
